# MGNREGA DATA ANALYSIS

## This dataset provides detailed insights into the implementation of the **Mahatma Gandhi National Rural Employment Guarantee Act (MGNREGA)** across various Indian states and districts from 2014 to 2022. It captures critical parameters such as:

- Number of job cards issued
- Households demanding and allotted work
- Person-days of employment generated (including gender-wise split)
- Financial disbursements (via banks and post offices)
- Participation of SC/ST communities
- District and state-level summaries

The dataset enables a deep dive into rural employment patterns, gaps in work allocation, and fund distribution efficiency — making it a powerful tool for both data-driven governance and academic exploration.

## Identifying Districts with the Highest Employment Gaps under MGNREGA

```python
df.columns
```

```python
df.columns.to_list()
```

```python
df_gap = df[['srcstatename','srcdistrictname','srcyear','households_that_demanded_work','households_that_were_allotted_work']].copy()
df_gap.isnull().sum()
df_gap.dropna(inplace=True)
```

```python
df_gap['employment_gap'] = df_gap['households_that_demanded_work'] - df_gap['households_that_were_allotted_work']
df_gap.head()
```

```python
# Top 10 districts with highest employment gap
top_gap_districts = df_gap.sort_values(by='employment_gap', ascending=False)
top_gap_districts
top_gap_districts = df_gap.sort_values(by='employment_gap', ascending=False).head(10)
top_gap_districts
```

### ✅ Conclusion:

The analysis reveals that **all top 10 districts with the highest employment gaps under MGNREGA belong to the state of Andhra Pradesh**. This indicates a significant mismatch between the number of households demanding employment and those actually allotted work in the region. 

Such consistent under-allotment suggests possible systemic inefficiencies in MGNREGA implementation at the district level in Andhra Pradesh. Further investigation into administrative, logistical, or budgetary constraints in these districts could help bridge the employment gap and ensure more equitable job distribution under the scheme.

```python
# Group by State and calculate total employment gap
state_gap = df_gap.groupby('srcstatename')['employment_gap'].sum().reset_index()
state_gap = state_gap.sort_values(by='employment_gap', ascending=False)
# heatmap using seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(state_gap.set_index('srcstatename'), annot=True, fmt=".0f", cmap='OrRd', linewidths=0.5, linecolor='gray')

plt.title('Total Employment Gap by State (MGNREGA)')
plt.ylabel('State')
plt.xlabel('Employment Gap')
plt.tight_layout()
plt.show()
```
